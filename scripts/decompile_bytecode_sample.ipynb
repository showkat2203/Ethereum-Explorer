```python
import pprint

apipath = "https://ethervm.io/decompile/"

# https://ethervm.io/decompile/0xcbe0db04cc90fa0e447d76c29c2695d27e02e058

address = "0xcbe0db04cc90fa0e447d76c29c2695d27e02e058"

new_api = apipath + address

s = requests.Session()
retries = Retry(total=5,
                backoff_factor=0.1,
                status_forcelist=[ 500, 502, 503, 504 ])

s.mount(new_api, HTTPAdapter(max_retries=retries))
response = s.get(new_api)

# dict_data = xmltodict.parse(response.content)
soup = BeautifulSoup(response.text, "html")

div = soup.find('div', {"class" : "code javascript"})

text = div.text
print(text)

# pprint.pprint(soup)
```
