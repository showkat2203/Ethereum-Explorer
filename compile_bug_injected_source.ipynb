```python
# import solcx

# from solcx import install_solc
# install_solc('v0.5.16')
# from solcx import get_installed_solc_versions
# from solcx import install_solc_pragma
# install_solc_pragma('^0.5.16')
# install_solc_pragma('^0.4.20 || >0.5.5 <0.7.0')
#install_solc('0.7.0')

bytecode_dict = dict()

for i in range(0, min(7000, len(file_list_full_path))):
#     print("File Name: ", file_list_full_path[i])

    sorted(contract_file_full_version[file_list_full_path[i]], reverse=True)
    
#     if( file_list_full_path[i] in skip_files ):
#         continue

#     if( len(contract_file_full_version[file_list_full_path[i]]) > 1):
#         print(i, " ****** ", file_list_full_path[i], "Avoided")
#         continue 
    cur_solc_version = contract_file_full_version[file_list_full_path[i]][0]
    if( cur_solc_version == "None" ):
        continue
    elif( len(cur_solc_version) == 3 ):
#         if( cur_solc_version == "0.5" ):
#             cur_solc_version += ".5"
#         else:
        cur_solc_version += ".0"
            
    try:
        res = solcx.compile_files( file_list_full_path[i], output_values=["abi", "bin-runtime"], solc_version=cur_solc_version)
    
    except:
        print("Error!!!!!  ", file_list_full_path[i])
        continue
    
    byte_code = ""
    cur_address = address_list[i]
    
    for key, value in res.items():
        temp = res[key]["bin-runtime"]
        temp = temp.replace("\n", "").replace(" ", "")
        byte_code +=  temp

    bytecode_dict[cur_address] = byte_code
    
#     break
# print("Done")
# pprint.pprint(bytecode_dict)

```
